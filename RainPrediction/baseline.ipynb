```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
train_end_idx = len(df_train) 
df_test = pd.read_csv('test.csv')
df_test['RainToday'] = np.zeros((len(df_test),))

# 為了處理方便，把 'train.csv' 和 'test.csv' 合併起來，'test.csv'的 RainToday 欄位用 0 補起來。
# 以 train_end_idx 作為 'train.csv' 和 'test.csv' 分界列，
df = pd.concat([df_train, df_test], sort=False)
```

```python
df.describe()
```

```python
# 將非數值欄位拿掉
#df = df.drop(columns = [col for col in df.columns if df[col].dtype == np.object])
df = df.drop(columns = 'Date')
#df = df.drop(columns = 'RISK_MM')
# # 將 missing value 補 0
# df = df.fillna(0)
```

```python
for col in df.columns:
    if(col != 'RainToday' and col != 'Location' and col != 'WindDir9am' and col != 'WindDir3pm' and col != 'WindGustDir'):
        col_median=df[col].median()
        df[col].fillna(col_median, inplace=True)  
    else:
        print('tala')
```

```python
for df2 in [df]:
    df2['WindGustDir'].fillna(df_train['WindGustDir'].mode()[0], inplace=True)
    df2['WindDir9am'].fillna(df_train['WindDir9am'].mode()[0], inplace=True)
    df2['WindDir3pm'].fillna(df_train['WindDir3pm'].mode()[0], inplace=True)
    df2['Location'].fillna(df_train['Location'].mode()[0], inplace=True)
```

```python
for cat_feature in ['Location','WindDir9am','WindDir3pm','WindGustDir']:
    df[cat_feature] = df[cat_feature].astype(str)
# one-hot encoding
df = pd.get_dummies(df)
```

```python
for col in df.columns:
    if(col != 'RainToday'):
        IQR = df[col].quantile(0.75) - df[col].quantile(0.25)
        Lower_fence = df[col].quantile(0.25) - (IQR * 1.5)
        Upper_fence = df[col].quantile(0.75) + (IQR * 1.5)
        print(col,'outliers are values < {lowerboundary} or > {upperboundary}'.format(lowerboundary=Lower_fence, upperboundary=Upper_fence))
```

```python
def max_value(df3, variable, top):
    return np.where(df3[variable]>top, top, df3[variable])

for df3 in [df]:
    df3['WindGustSpeed'] = max_value(df3, 'WindGustSpeed', 58.0)
    df3['Evaporation'] = max_value(df3, 'Evaporation', 6.14)
    df3['WindSpeed9am'] = max_value(df3, 'WindSpeed9am', 29.0)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
sc.fit(df)
df = pd.DataFrame(sc.transform(df), index=df.index, columns=df.columns)
```

```python
df.describe()
```

# train

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns = ['RainToday']).values[:train_end_idx, :],
    df['RainToday'].values[:train_end_idx], test_size=0.5)
X_ans = df.drop(columns = ['RainToday']).values[train_end_idx:, :]
```

```python
import xgboost
params = {}
params['objective'] = 'binary:logistic'
params['eval_metric'] = 'logloss'
params['eta'] = 0.04
params['max_depth'] = 8
params['learning_rate'] = 0.01

d_train = xgboost.DMatrix(X_train, label=y_train)
d_valid = xgboost.DMatrix(X_test, label=y_test)

watchlist = [(d_train, 'train'), (d_valid, 'valid')]

bst = xgboost.train(params, d_train, 1000, watchlist, early_stopping_rounds=100, verbose_eval=10)
y_pred = bst.predict(xgboost.DMatrix(X_test))
```

```python
from sklearn.metrics import accuracy_score, f1_score
y_pred_decision = [0] * len(y_pred)
for i in range(len(y_pred)):
    if(y_pred[i] <= 0.19):
        y_pred_decision[i] = 0
    else:
        y_pred_decision[i] = 1
print('Accuracy: %f' % accuracy_score(y_test, y_pred_decision))
print('f1-score: %f' % f1_score(y_test, y_pred_decision))
```

```python
ans_pred = bst.predict(xgboost.DMatrix(X_ans))
for i in range(len(ans_pred)):
    if(ans_pred[i] <= 0.19):
        ans_pred[i] = 0
    else:
        ans_pred[i] = 1
df_sap = pd.DataFrame(ans_pred.astype(int), columns = ['RainToday'])
df_sap.to_csv('myAns.csv',  index_label = 'Id')
```

```python
from imblearn.under_sampling import EditedNearestNeighbours
enn = EditedNearestNeighbours()
X_train, y_train = enn.fit_resample(X_train, y_train)
```

```python
from imblearn.over_sampling import SMOTE, ADASYN
X_train, y_train = SMOTE().fit_resample(X_train, y_train)
```

```python
from imblearn.over_sampling import BorderlineSMOTE
X_train, y_train = BorderlineSMOTE().fit_resample(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score

model = DecisionTreeClassifier()
model.fit(X_train,y_train)

#predict
y_pred_decision = model.predict(X_test)
print('Accuracy: %f' % accuracy_score(y_test, y_pred_decision))
print('f1-score: %f' % f1_score(y_test, y_pred_decision))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(criterion='gini',
                                n_estimators=20,random_state=1)
model.fit(X_train,y_train)
y_pred_decision = model.predict(X_test)
print('Accuracy: %f' % accuracy_score(y_test, y_pred_decision))
print('f1-score: %f' % f1_score(y_test, y_pred_decision))
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model  import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred_decision = model.predict(X_test)
print('Accuracy: %f' % accuracy_score(y_test, y_pred_decision))
print('f1-score: %f' % f1_score(y_test, y_pred_decision))
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(StandardScaler(), PCA(n_components=2), GaussianNB())
model.fit(X_train, y_train)
y_pred_decision = model.predict(X_test)
print('Accuracy: %f' % accuracy_score(y_test, y_pred_decision))
print('f1-score: %f' % f1_score(y_test, y_pred_decision))
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model  import LogisticRegression
model = LogisticRegression(class_weight="balanced")
model.fit(X_train, y_train)
mythreshold = 0.69
y_pred_decision = np.where(model.predict_proba(X_test)[:,1] > mythreshold, 1, 0)
print('Accuracy: %f' % accuracy_score(y_test, y_pred_decision))
print('f1-score: %f' % f1_score(y_test, y_pred_decision))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_decision)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```
